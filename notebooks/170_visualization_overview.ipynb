# Data Visualization Overview

The landscape of visualization libraries that you can use in Python is very broad and it is very difficult to select the best tool
for your needs (on the other hand, it can be aesthetically pleasing to experiment with them :-)). The following image illustrates the situation:

![Python Visualization Landscape](images/vis-landscape.png)

(taken from https://github.com/rougier/python-visualization-landscape)

## Assorted list of visualization libraries

- [matplotlib](https://matplotlib.org/) - mature library for (mostly) static plots
- [seaborn](https://seaborn.pydata.org/) - high-level interface for drawing attractive and informative statistical graphics (using matplotlib)
- [bokeh](https://docs.bokeh.org/en/latest/index.html) - interactive visualization library for creating  interactive plots, dashboards, and data applications.
- [datashader](https://datashader.org/) - a graphics pipeline system for creating meaningful representations of large datasets quickly and flexibly
- [holoviews](https://holoviews.org/) - high-level library that abstracts the library used (matplotlib / bokeh /...) and focuses on sematics.
- [altair](https://altair-viz.github.io/) - declarative statistical visualization library for Python, based on [vega](https://vega.github.io/vega/) and vega-lite plotting grammars.
- [plotnine](https://plotnine.readthedocs.io/en/stable/) - implementation of a grammar of graphics in Python, similar to R's ggplot2
- [plotly](https://plotly.com/python/) - broadly used interactive plotting library (plotly.express)
- [d3.js](https://d3js.org/) - a JavaScript library for manipulating documents based on data (not to be used directly in Python, but as a frontend tool)
